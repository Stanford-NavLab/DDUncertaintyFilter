```python
def calc_single_diff(meas_a, meas_b):
    return meas_a - meas_b

def calc_double_diff(meas_ac, meas_bc, meas_ad, meas_bd):
    c_sd = calc_single_diff(meas_ac, meas_bc)
    d_sd = calc_single_diff(meas_ad, meas_bd)
    t_max = len(d_sd)
    
    c_dd = np.zeros((t_max, c_sd.shape[1]))
    d_dd = np.zeros((t_max, d_sd.shape[1]))
    for t in range(t_max):
        if not np.all(np.isnan(c_sd[t, :])):
            refsat = np.nanargmin(c_sd[t, :])
            c_dd[t, :] = c_sd[t, :] - c_sd[t, refsat]
            d_dd[t, :] = d_sd[t, :] - c_sd[t, refsat]
            
    c_dd[c_dd==0] = np.nan
    d_dd[d_dd==0] = np.nan
    
    return c_dd, d_dd
```

```python
def calc_residuals(t_gt, ecef_rec, const, rec):
    if const=='gps':
        svs_all = list(gps_eph.sv)
        if rec=='rover':
            code = code_rg
            carrier = carrier_rg
        elif rec=='base':
            code = code_bg
            carrier = carrier_bg
        svs = svs_gps
        eph = gps_eph
        t_obs = t_gps
        lda = lda_gps
    elif const=='beidou':
        svs_all = list(bei_eph.sv)
        if rec=='rover':
            code = code_rb
            carrier = carrier_rb
        elif rec=='base':
            code = code_bb
            carrier = carrier_bb
            
        svs = svs_bei
        eph = bei_eph
        t_obs = t_bei
        lda = lda_bei
    else:
        raise("Not implemented")
        
    if len(ecef_rec.shape)==1:
        ecef_rec = np.tile(ecef_rec, (len(t_gt), 1))
    
    sv2idx = lambda svs: [svs_all.index(sv) for sv in svs]
    tidx_gt2meas = np.abs((np.array(t_obs)[:, None] - t_gt)).argmin(axis=0)
    
    residuals_code = np.zeros((len(t_gt), len(svs_all)))
    residuals_carr = np.zeros((len(t_gt), len(svs_all)))

    for idx in tqdm(range(len(t_gt))):
        meas_idx = tidx_gt2meas[idx]
        
        svs_filtered = [sv for sv in svs[meas_idx] if sv in svs_all]
        code_filtered = np.array([code[meas_idx][i] for i, sv in enumerate(svs[meas_idx]) if sv in svs_all])
        carrier_filtered = np.array([carrier[meas_idx][i] for i, sv in enumerate(svs[meas_idx]) if sv in svs_all])
#         svs_mask = [i for i, sv in enumerate(svs[meas_idx]) if sv in svs_all]
        if len(svs_filtered) == 0:
            continue
        flight_times = computeFlightTimes(code_filtered, svs_filtered, eph, t_gt[idx])
        ecef_svs = getPos(eph, t_gt[idx], flight_times, svs_filtered)
        
#         print(t_gt[idx], t_obs[meas_idx], ecef_svs, code_filtered, carrier_filtered)
#         print("--------------------------------------------------")
        
        residuals_code[idx, sv2idx(svs_filtered)] = code_filtered - np.linalg.norm(ecef_svs - ecef_rec[idx, None, :], axis=1)
        residuals_carr[idx, sv2idx(svs_filtered)] = lda*carrier_filtered - np.linalg.norm(ecef_svs - ecef_rec[idx, None, :], axis=1)
        
    residuals_code[residuals_code==0] = np.nan
    residuals_carr[residuals_carr==0] = np.nan
    return residuals_code, residuals_carr
```

```python
def create_observables(t_gt, ecef_rec, const, rec):
    if const=='gps':
        svs_all = list(gps_eph.sv)
        if rec=='rover':
            code = code_rg
            carrier = carrier_rg
            cnos = cnos_rg
        elif rec=='base':
            code = code_bg
            carrier = carrier_bg
            cnos = None
        svs = svs_gps
        eph = gps_eph
        t_obs = t_gps
        lda = lda_gps
    elif const=='beidou':
        svs_all = list(bei_eph.sv)
        if rec=='rover':
            code = code_rb
            carrier = carrier_rb
            cnos = cnos_rb
        elif rec=='base':
            code = code_bb
            carrier = carrier_bb
            cnos = None
        svs = svs_bei
        eph = bei_eph
        t_obs = t_bei
        lda = lda_bei
    else:
        raise("Not implemented")
        
    if len(ecef_rec.shape)==1:
        ecef_rec = np.tile(ecef_rec, (len(t_gt), 1))
    
    sv2idx = lambda svs: [svs_all.index(sv) for sv in svs]
    tidx_gt2meas = np.abs((np.array(t_obs)[:, None] - t_gt)).argmin(axis=0)
    
    measurements_code = np.zeros((len(t_gt), len(svs_all)))
    measurements_carr = np.zeros((len(t_gt), len(svs_all)))
    measurements_cnos = np.zeros((len(t_gt), len(svs_all)))
    ecef_svs_all = np.zeros((len(t_gt), len(svs_all), 3))

    for idx in tqdm(range(len(t_gt))):
        meas_idx = tidx_gt2meas[idx]
        
        svs_filtered = [sv for sv in svs[meas_idx] if sv in svs_all]
        code_filtered = np.array([code[meas_idx][i] for i, sv in enumerate(svs[meas_idx]) if sv in svs_all])
        carrier_filtered = np.array([carrier[meas_idx][i] for i, sv in enumerate(svs[meas_idx]) if sv in svs_all])
        if cnos is not None:
            cnos_filtered = np.array([cnos[meas_idx][i] for i, sv in enumerate(svs[meas_idx]) if sv in svs_all])
        else:
            cnos_filtered = None
#         svs_mask = [i for i, sv in enumerate(svs[meas_idx]) if sv in svs_all]
        if len(svs_filtered) == 0:
            continue
        flight_times = computeFlightTimes(code_filtered, svs_filtered, eph, t_gt[idx])
        ecef_svs = getPos(eph, t_gt[idx], flight_times, svs_filtered)
        
#         print(t_gt[idx], t_obs[meas_idx], ecef_svs, code_filtered, carrier_filtered)
#         print("--------------------------------------------------")
        
        measurements_code[idx, sv2idx(svs_filtered)] = code_filtered
        measurements_carr[idx, sv2idx(svs_filtered)] = lda*carrier_filtered
        if cnos_filtered is not None:
            measurements_cnos[idx, sv2idx(svs_filtered)] = cnos_filtered
        ecef_svs_all[idx, sv2idx(svs_filtered), :] = ecef_svs
        
    measurements_code[measurements_code==0] = np.nan
    measurements_carr[measurements_carr==0] = np.nan
    measurements_cnos[measurements_cnos==0] = np.nan
    print("Saving...")
    np.save("save_data/"+const+"_"+rec+"_measurements_code_new.npy", measurements_code)
    np.save("save_data/"+const+"_"+rec+"_measurements_carr_new.npy", measurements_carr)
    np.save("save_data/"+const+"_"+rec+"_measurements_cnos_new.npy", measurements_cnos)
    np.save("save_data/"+const+"_ecef_svs_new.npy", ecef_svs_all)
    np.save("save_data/time_gt_new.npy", t_gt)
```
